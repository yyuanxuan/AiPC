```python
####################################################################################################
# Ancillary 
# Using Clustergram to determin the number of groups (sub-clusters) in each supergroup clusters
####################################################################################################
import pandas as pd 
from clustergram import Clustergram
import matplotlib.pyplot as plt
```

```python
# Import/read data (pre-processed) of all LSOA in England
df= pd.read_csv('../data/input/input-data-boxcox-scale.csv')
df
```

```python
# Create clustergram for all LSOAs in England
# Determin the number of Supergroups (first-tier clusters) in AiPC
cgram = Clustergram(range(2, 13), backend= 'sklearn', method='kmeans', n_init=1000, random_state=123, max_iter=100000,init='k-means++')
cgram.fit(df.drop(columns = ['geography_code']))
cgram.plot(
    size=0.8,
    linewidth=1,
    cluster_style={"color": "#ffeead", "edgecolor": "#de4307"},
    line_style={"color": "#ff8a5c"},
    figsize=(15, 6)
    )

plt.savefig('../img/clustergram-supergroup.svg')
plt.savefig('../img/clustergram-supergroup.png', dpi=300)
```
